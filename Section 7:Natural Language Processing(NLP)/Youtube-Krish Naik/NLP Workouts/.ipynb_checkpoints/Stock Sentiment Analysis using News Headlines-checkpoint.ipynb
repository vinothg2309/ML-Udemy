```python
import pandas as pd
```

```python
df = pd.read_csv("stock_data.csv", encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
train = df[df['Date'] < '20071231']
test = df[df['Date'] > '20080101']
```

```python
data = train.iloc[:,2:]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)

#Renaming column names for ease of use
new_index = [str(i) for i in range(25)]
data.columns = new_index
data.head()
```

```python
for index in new_index:
    data[index] = data[index].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[0,:])
```

```python
# To apply BOW,TF-IDF or Word2Vec, we need all data in a paragraph. Doing so, it can create BOW.
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,:]))
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

An n-gram is a contiguous sequence of n words, for example, in the sentence "dog that barks does not bite", the n-grams are:
      unigrams (n=1):Â dog, that, barks, does, not, bite
    â€¢ bigrams (n=2):Â dog that, that barks, barks does, does not, not bite
    â€¢ trigrams (n=3):Â dog that barks, that barks does, barks does not, does not bite

```python
countvector=CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headlines)
```

```python
#Implement randomforest classifier
randomclassifier = RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
print(train['Label'])
```

```python
# To apply BOW,TF-IDF or Word2Vec, we need all data in a paragraph. Doing so, it can create BOW.
test_headlines = []
for row in range(0,len(test.index)):
    test_headlines.append(' '.join(str(x) for x in data.iloc[row,:]))
test_dataset = countvector.transform(test_headlines)
predictions = randomclassifier.predict(test_dataset)
```

```python
print(test['Label'])
print(test.head())
```

```python
from sklearn.metrics import classification_report, accuracy_score
#print(classification_report(test['Label'],predictions))
print('accuracy_score : ',accuracy_score(test['Label'],predictions))
```
