```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Section 3:Classification/Logistic Regression/Social_Network_Ads.csv")
print(dataset.head())
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1:].values
print(X)
print(y)
```

# Splitting dataset to Training and test set

On changing random_state, we will get different result in confusion_matrix, accuracy_score

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

# Featuring scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

# Training K-NN model on training data

```python
# KNN calculates euclidean distance between new data with trained data to classify new incoming data
from sklearn.neighbors import KNeighborsClassifier
# n_neighbors-No of neighbors to consider(default:5),weights 
classifier=KNeighborsClassifier(n_neighbors=5, metric='minkowski',p=2)
classifier.fit(X_train,np.ravel(y_train))
y_pred=classifier.predict(X_test)
print(classifier.predict(sc.transform([[30,87000]])))
print(classifier.predict(sc.transform([[36,144000]])))
```

```python
print(type(y_pred),type(y_test))
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
# Row - Actual Value, column - Predicted value
#   0       1(Predicted)
#0 50(c)    8
#1  4      38(C) # C - Correct Prediction
accuracy_score(y_test,y_pred)
tn, fp, fn, tp = cm.ravel()
print(tn, fp, fn, tp)
precision_0 = tn/(tn+fn)
print('precision_0 : ', precision_0)
```

```python
print(confusion_matrix([0, 1, 0, 1], [1, 1, 1, 0]))
tn, fp, fn, tp = confusion_matrix([0, 1, 0, 1], [1, 1, 1, 0]).ravel()
print(tn, fp, fn, tp)
tn, fp, fn, tp = confusion_matrix([0, 1, 0, 1], [1, 1, 1, 0]).reshape(-1)
print(tn, fp, fn, tp)
```

```python
from sklearn.metrics import classification_report
cp = classification_report(y_test,y_pred)
print(cp)
```

# Cross Validation

# K Fold CV

ravel - returns array with same memory reference as original array shape (n_samples,). Changing value with refect changes in original array
flatten - returns copy of original array(shape (n_samples,1)). Changing value won't affect original array

```python
from sklearn.model_selection import cross_val_score
# its creates 10 folds and apply prediction on all 10 folds. 
# If 1000 records are available in dataset, each fold contains 100 records(1000/10)
# 1st iteration-1st fold(0-100) is test set & rest is training set. 
# 2nd iteration-2nd fold(100-200) as testset & restis training set.
score=cross_val_score(classifier,X,y.ravel(),cv=10) 
print('Score : ', score)
print('Avg. Score : ', np.mean(score))
print(np.shape(y))
print(np.shape(y.ravel()))
```

# K fold cross validation

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
accuracy=[]
kf=KFold(n_splits=10) # it won't shuffle the index
for train_index, test_index in kf.split(X,np.ravel(y)):
    print("Train Index : ", train_index, " Test Index : ", test_index)
    Xf_train,Xf_test=X[train_index], X[test_index]
    yf_train,yf_test=y[train_index], y[test_index]
    print("X1_train : ", len(Xf_train), " X1_test : ", len(Xf_test), " y1_train : ", len(yf_train)," y1_test : ", len(yf_test))
    classifier.fit(Xf_train,yf_train.ravel())
    prediction=classifier.predict(Xf_test)
    score=accuracy_score(prediction,yf_test)
    accuracy.append(score)

print("Accuracy : ", accuracy)
print("Avg. Accuracy : ", np.mean(accuracy))    
```

# Stratified K fold cross validation

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
accuracy=[]
skf=StratifiedKFold(n_splits=10) # Shuffle train and test set
for train_index, test_index in skf.split(X,np.ravel(y)):
    print("Train Index : ", train_index, " Test Index : ", test_index)
    X1_train,X1_test=X[train_index], X[test_index]
    y1_train,y1_test=y[train_index], y[test_index]
    print("X1_train : ", len(X1_train), " X1_test : ", len(X1_test), " y1_train : ", len(y1_train)," y1_test : ", len(y1_test))
    classifier.fit(X1_train,y1_train.ravel())
    prediction=classifier.predict(X1_test)
    score=accuracy_score(prediction,y1_test)
    accuracy.append(score)

print("Accuracy : ", accuracy)
print("Avg. Accuracy : ", np.mean(accuracy))    
```

```python
from sklearn.model_selection import KFold, StratifiedKFold, StratifiedShuffleSplit
SEED = 43
SPLIT = 3
X_train = [0,1,2,3,4,5,6,7,8]
y_train = [0,0,0,0,0,0,1,1,1]   # note 6,7,8 are labelled class '1'

print("KFold, shuffle=False (default)")
kf = KFold(n_splits=SPLIT, random_state=SEED)
for train_index, test_index in kf.split(X_train, y_train):
    print("TRAIN:", train_index, "TEST:", test_index)

print("KFold, shuffle=True")
kf = KFold(n_splits=SPLIT, shuffle=True, random_state=SEED)
for train_index, test_index in kf.split(X_train, y_train):
    print("TRAIN:", train_index, "TEST:", test_index)

print("\nStratifiedKFold, shuffle=False (default)")
skf = StratifiedKFold(n_splits=SPLIT, random_state=SEED)
for train_index, test_index in skf.split(X_train, y_train):
    print("TRAIN:", train_index, "TEST:", test_index)
    
print("StratifiedKFold, shuffle=True")
skf = StratifiedKFold(n_splits=SPLIT, shuffle=True, random_state=SEED)
for train_index, test_index in skf.split(X_train, y_train):
    print("TRAIN:", train_index, "TEST:", test_index)
    
print("\nStratifiedShuffleSplit")
sss = StratifiedShuffleSplit(n_splits=SPLIT, random_state=SEED, test_size=3)
for train_index, test_index in sss.split(X_train, y_train):
    print("TRAIN:", train_index, "TEST:", test_index)

print("\nStratifiedShuffleSplit (can customise test_size)")
sss = StratifiedShuffleSplit(n_splits=SPLIT, random_state=SEED, test_size=2)
for train_index, test_index in sss.split(X_train, y_train):
    print("TRAIN:", train_index, "TEST:", test_index)
```
