If there are N items that need to be classified into C mutually exclusive categories, the work of Cohen's kappa is to **measure the agreement between the two raters in order to classify N to C.**

**It basically tells you how much better your classifier is performing over the performance of a classifier that simply guesses at random according to the frequency of each class.**

**KappaScore=(Agree-ChanceAgree)/(1-ChanceAgree)**

If Agree=1, we have perfect agreement, corresponding to a matrix where all the non-diagonal (pink) cells are 0. In this case, the kappa score is 1, regardless of ChanceAgree. In contrast, if Agree=ChanceAgree, kappa is 0, signifying that the professors' agreement is by chance. If Agree is smaller than ChanceAgree, the kappa score is negative, denoting that the degree of agreement is lower than chance agreement.

According to their scheme a value < 0 is indicating no agreement , 0–0.20 as slight, 0.21–0.40 as fair, 0.41–0.60 as moderate, 0.61–0.80 as substantial, and 0.81–1 as almost perfect agreement.

```python
r1=['yes','no','yes','no','yes','no','yes','no','yes']

r2=['yes','yes','yes','no','no','no','yes','yes','yes']
```

```python
from sklearn.metrics import confusion_matrix, cohen_kappa_score
cm = confusion_matrix(r1,r2)
print(cm)
print(cohen_kappa_score(r1,r2))
```

### Kappa Score Manual calculation







```python
tn,fp,fn,tp = cm.reshape(-1)
print('tn : {} ,fp : {},fn : {},tp :{}'.format(tn,fp,fn,tp))
total_size = tn+fp+fn+tp
```

```python
Agree = (tp+tn)/total_size
chanceAgree_0 = ((tn+fp)/total_size)*((tn+fn)/total_size)
chanceAgree_1 = ((fn+tp)/total_size)*((fp+tp)/total_size)
print('chanceAgree_0 : ', chanceAgree_0, ' chanceAgree_1 : ', chanceAgree_1)
chance_agree = chanceAgree_0+chanceAgree_1
kappa_score = (Agree-chance_agree)/(1-chance_agree)
print('kappa_score : ', kappa_score)
```
