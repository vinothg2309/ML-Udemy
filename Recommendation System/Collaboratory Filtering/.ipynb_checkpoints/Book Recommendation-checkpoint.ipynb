### **To ensure statistical significance, users with less than 200 ratings, and books with less than 100 ratings are excluded.**

```python
counts1 = ratings['userID'].value_counts()
ratings = ratings[ratings['userID'].isin(counts1[counts1 >= 200].index)]
counts = ratings['bookRating'].value_counts()
ratings = ratings[ratings['bookRating'].isin(counts[counts >= 100].index)]
```

```python
ratings.head()
```

### **Collaborative Filtering Using k-Nearest Neighbors (kNN)**

kNN is a machine learning algorithm to find clusters of similar users based on common book ratings, and make predictions using the average rating of top-k nearest neighbors. For example, we first present ratings in a matrix with the matrix having one row for each item (book) and one column for each user,

```python
combine_book_rating = pd.merge(ratings, books, on='ISBN')
columns = ['yearOfPublication', 'publisher', 'bookAuthor', 'imageUrlS', 'imageUrlM', 'imageUrlL']
combine_book_rating = combine_book_rating.drop(columns, axis=1)
combine_book_rating.head()
```

We then group by book titles and create a new column for total rating count.

```python
combine_book_rating = combine_book_rating.dropna(axis=0, subset=['ISBN'])

book_ratingCount = combine_book_rating.groupby('bookTitle')['bookRating'].count().reset_index().rename(columns = {'bookRating': 'totalRatingCount'})[['bookTitle', 'totalRatingCount']]
book_ratingCount.head()
```

```python
rating_with_totalRatingCount = pd.merge(combine_book_rating,book_ratingCount,on='bookTitle')
rating_with_totalRatingCount.head()
```

```python
popularity_threshold = 50
rating_popular_book = rating_with_totalRatingCount.query("totalRatingCount > @popularity_threshold")
rating_popular_book.head()
```

```python
rating_popular_book.shape
```

### **Filter to users in USA & Canada only**
KNN is very slow if v consider whole dataset, so filtring only USA & Canada customer

```python
users.head()
```

```python
combined = pd.merge(rating_popular_book,users, on="userID")
us_canada_user_rating = combined[combined['Location'].str.contains('usa|canada',na=False,case=False)]
us_canada_user_rating.drop('Age',axis=1,inplace=True)
us_canada_user_rating.head()
```

Cos 0 =1, cos 90 = 0



### Implementing kNN
We convert our table to a 2D matrix, and fill the missing values with zeros (since we will calculate distances between rating vectors). We then transform the values(ratings) of the matrix dataframe into a scipy sparse matrix for more efficient calculations.

Finding the Nearest Neighbors We use unsupervised algorithms with sklearn.neighbors. The algorithm we use to compute the nearest neighbors is “brute”, and we specify “metric=cosine” so that the algorithm will calculate the cosine similarity between rating vectors. Finally, we fit the model.

```python
from scipy.sparse import csr_matrix
us_canada_user_rating = combined.drop_duplicates(['userID','bookTitle'])
us_canada_user_pivot = us_canada_user_rating.pivot(index='bookTitle',columns='userID',values='bookRating').fillna(0)
#print(us_canada_user_pivot.head())
us_canada_user_matrix = csr_matrix(us_canada_user_pivot.values)
print('us_canada_user_matrix :\n', us_canada_user_matrix)

from sklearn.neighbors import NearestNeighbors
model_knn=NearestNeighbors(metric='cosine',algorithm='brute')
model_knn.fit(us_canada_user_matrix)
```

```python
us_canada_user_pivot.head()
```

```python
#print(us_canada_user_pivot.shape())
query_index = np.random.choice(us_canada_user_pivot.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(us_canada_user_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
print(distances)
print(indices)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(us_canada_user_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, us_canada_user_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```
